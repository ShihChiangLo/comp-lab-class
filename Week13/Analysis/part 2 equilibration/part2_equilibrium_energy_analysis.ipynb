```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mdtraj as md
import matplotlib
from matplotlib.animation import FFMpegWriter, PillowWriter
```

```python
#data from ener
file = pd.read_csv('./NVT-1.ener',skiprows=1,sep='\s+',header=None)
file.columns = ['Step','Time','Kin','Temp','Pot','Con Qty','Used time']
NPT_file = pd.read_csv('./NPT-1.ener',skiprows=1,sep='\s+',header=None)
NPT_file.columns = ['Step','Time','Kin','Temp','Pot','Con Qty','Used time']
```

```python
plt.plot(file.Time, file.Pot,label='potential energy')
plt.xlabel('Time (fs)')
plt.ylabel('Potential energy (a.u.)')
plt.title('NVT Potential energy vs Time')
plt.legend()
plt.savefig('NVT Potential energy vs Time.pdf')
plt.show()
```

```python
plt.plot(NPT_file.Time, NPT_file.Pot,label='potential energy')
plt.xlabel('Time (fs)')
plt.ylabel('Potential energy (a.u.)')
plt.title('NPT Potential energy vs Time')
plt.legend()
plt.savefig('NPT Potential energy vs Time.pdf')
plt.show()
```
