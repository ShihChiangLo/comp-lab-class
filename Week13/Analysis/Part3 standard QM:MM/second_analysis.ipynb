```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mdtraj as md
import matplotlib
from matplotlib.animation import FFMpegWriter, PillowWriter
```

```python
#data from monitor ener
file = pd.read_csv('./MONITOR-1.ener',skiprows=1,sep='\s+',header=None)
file.columns = ['Step','Time','Kin','Temp','Pot','Con Qty','Used time']
```

```python
plt.plot(file.Time, file.Pot,label='potential energy')
plt.xlabel('Time (fs)')
plt.ylabel('Potential energy (a.u.)')
plt.title('Monitor_Potential energy vs Time')
plt.legend()
plt.savefig('Monitor_Potential energy vs Time.pdf')
plt.show()
```

```python
#data from monitor reaction coordinate
reaction_coordinate_file = pd.read_csv('./MONITOR-COLVAR.metadynLog',skiprows=1,sep='\s+',header=None)
reaction_coordinate_file.columns = ['Time','Energy','X','Y','z','K']
```

```python
plt.plot(reaction_coordinate_file.Time, reaction_coordinate_file.Energy,label='energy')
plt.xlabel('reaction')
plt.ylabel('d(CC)-d(OC)')
plt.title('Reaction coordinate')
plt.legend()
plt.savefig('Reaction coordinate.pdf')
plt.show()
```
