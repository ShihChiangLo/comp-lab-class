```python
datasets = [u_15, u_10, u_09, u_08, u_07, u_065, u_06, u_055, u_05, u_0475]
for i in range(len(temp)):
    MSD = msd.EinsteinMSD(datasets[i], select = 'type 1', fft = True)
    MSD.run()
    msdresults = MSD.results.timeseries
    nframes = MSD.n_frames
    timestep = 1000000
    lagtimes = np.arange(nframes)*timestep
    plt.loglog(lagtimes, msdresults, label = 'T='+str(temp[i]))
plt.legend()
plt.xlabel("log time")
plt.ylabel("log MSD")
plt.title('MSD of Type 1 Particles')
plt.tight_layout()
#plt.savefig('MSD_1.pdf', format='pdf', dpi=300)
plt.show()
```

```python
zeropoint = []
for i in range(len(temp)):
    MSD = msd.EinsteinMSD(datasets[i], select = 'type 1', fft = True)
    MSD.run()
    msdresults = MSD.results.timeseries
    nframes = MSD.n_frames
    timestep = 1000000
    lagtimes = np.arange(nframes)*timestep
    zeropoint.append(np.interp(1, msdresults, lagtimes))
plt.loglog(zeropoint, 1/np.array(temp))
plt.xlabel("log time at MSD = 1 of different temperatures")
plt.ylabel("log 1/T")
plt.title('Log plot of time vs 1/T')
plt.tight_layout()
#plt.savefig('plot of MSD equals to one at different temperature.pdf', format='pdf', dpi=300)
plt.show()
```
