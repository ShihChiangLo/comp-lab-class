```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import pandas as pd
```

```python
#import xvg content into jupyter notebook
with open("potential.xvg") as f:#potential
    potential = f.read()
    potential = potential.split('\n')
with open("temperature.xvg") as f:
    temperature = f.read()
    temperature = temperature.split('\n')
with open("pressure.xvg") as f:
    pressure = f.read()
    pressure = pressure.split('\n')
with open("density.xvg") as f:
    density = f.read()
    density = density.split('\n')
with open("rmsd.xvg") as f:
    rmsd = f.read()
    rmsd = rmsd.split('\n')
with open("rmsd_xtal.xvg") as f:
    rmsd_xtal = f.read()
    rmsd_xtal = rmsd_xtal.split('\n')
with open("gyrate.xvg") as f:
    gyration = f.read()
    gyration = gyration.split('\n')
```

```python
#Energy Minimization
#data cleaning
t = []#time
p = []#potential energy
for line in potential[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    p.append(float(x[1]))

#plot
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_title("Energy Minimization")    
ax1.set_xlabel('Time (ps)')
ax1.set_ylabel('potential energy (KJ/mol)')
ax1.plot(t,p, c='r', label='potenital')
leg = ax1.legend()
#plt.savefig('Energy Minimization.png')
plt.show() 
```

```python
#Temperature during Equilibration
#data cleaning
t = []#time
temp = []#temperature
for line in temperature[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    temp.append(float(x[1]))
    
#plot
fig = plt.figure()
ax2 = fig.add_subplot(111)
ax2.set_title(r'Temperature ')    
ax2.set_xlabel('Time (ps)')
ax2.set_ylabel('Temperature (K)')
ax2.plot(t,temp, c='b', label='Temperature')
leg = ax2.legend()
#plt.savefig('Temperature.png')
plt.show() 
```

```python
#Pressure during Equilibration
#data cleaning
t = []#time
press = []#pressure
for line in pressure[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    press.append(float(x[1]))

#rolling average 
window_size = 10
num_series = pd.Series(press)
windows = num_series.rolling(window_size)
moving_averages = windows.mean()
moving_averages_list = moving_averages.tolist()
final_list = moving_averages_list[window_size-1:]

#plot
fig = plt.figure()
ax3 = fig.add_subplot(111)
ax3.set_title(r'Pressure')    
ax3.set_xlabel('Time (ps)')
ax3.set_ylabel('pressure(bar)')
ax3.plot(t,press, c='b', label='Pressure')
ax3.plot(final_list, c='r', label='10-ps Running average')
leg = ax3.legend()
#plt.savefig('Pressure.png')
plt.show() 
```

```python
#Density during Equilibration
#data cleaning
t = []#time
dens = []#density
for line in density[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    dens.append(float(x[1]))

#rolling average 
window_size = 10
num_series = pd.Series(dens)
print(num_series)
windows = num_series.rolling(window_size)
moving_averages = windows.mean()
moving_averages_list = moving_averages.tolist()
final_list = moving_averages_list[window_size-1:]

#plot
fig = plt.figure()
ax4 = fig.add_subplot(111)
ax4.set_title(r'Density')    
ax4.set_xlabel('Time (ps)')
ax4.set_ylabel('Density(kg/m^3)')
ax4.plot(t,dens, c='b', label='Density')
ax4.plot(final_list, c='r', label='10-ps Running average')
leg = ax4.legend()
#plt.savefig('Density.png')
plt.show() 
```

```python
#rmsd_xtal and rmsd
#data cleaning
t = []#time
rmsd_equilibrated = []#rmsd equilibrated
rmsd_crystal = []#rmsd crystal

for line in rmsd[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    rmsd_equilibrated.append(float(x[1]))
    
for line in rmsd_xtal[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    rmsd_crystal.append(float(x[1]))

#plot
fig = plt.figure()
ax5 = fig.add_subplot(111)
ax5.set_title(r'RMSD')    
ax5.set_xlabel('Time (ns)')
ax5.set_ylabel('RMSD(nm)')
ax5.plot(t,rmsd_equilibrated, c='k', label='Ref: Equilibrated')
ax5.plot(t,rmsd_crystal, c='r', label='Ref: Crystal')
leg = ax5.legend()
#plt.savefig('RMSD.png')
plt.show() 
```

```python
#Radius of gyration
#data cleaning
t = []#time
Gy = []#gyration
for line in gyration[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    t.append(float(x[0]))
    Gy.append(float(x[1]))
    
#plot
fig = plt.figure()
ax6 = fig.add_subplot(111)
ax6.set_title(r'Radius of gyration')    
ax6.set_xlabel('Time (ps)')
ax6.set_ylabel('Rg(nm)')
ax6.plot(t,Gy, c='k')
plt.ylim([1.3,1.5])
#plt.savefig('Radius of gyration.png')
plt.show() 
```
