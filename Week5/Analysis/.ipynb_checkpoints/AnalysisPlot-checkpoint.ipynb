```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import pandas as pd
```

```python
#Trajectory load from trr and gro file, then compute phi and psi
trajectory = md.load("adp_T300_5ns.trr",top="adp_T300_5ns.gro")
phi_tem = md.compute_phi(trajectory,periodic=True,opt=True)
psi_tem = md.compute_psi(trajectory,periodic=True,opt=True)

phi = []
psi = []

for i in phi_tem[1]:
    phi.append(i[0])
    
for i in psi_tem[1]:
    psi.append(i[0])
```

```python
#plot phi and psi versus time

#phi vs time
plt.plot(trajectory.time, phi)
plt.title('phi vs time')
plt.xlabel('Time(ps)')
plt.ylabel('phi')
#plt.savefig('phi vs time.pdf')
plt.show()

#psi vs time
plt.plot(trajectory.time, psi)
plt.title('psi vs time')
plt.xlabel('Time(ps)')
plt.ylabel('psi')
#plt.savefig('psi vs time.pdf')
plt.show()
```

```python
#plot 2d histogram of phi and psi
His = np.histogram2d(phi,psi,bins=20)
#-log(histogram)
minus_log_His = -np.log(His[0])
plt.pcolormesh(minus_log_His)
plt.title('Free energy surface of psi and phi')
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.colorbar()
#plt.savefig('Free energy surface of psi and phi.pdf')
```

```python
#demux plot
#load xvg file
with open("replica_temp.xvg") as f:
    replica = f.read()
    replica = replica.split('\n')
    replica = replica[:-1]
#load data of different replicas    
replica_0=[]
replica_1=[]
replica_2=[]
replica_3=[]
time = np.arange(0,5000,10)
sampling=np.arange(0,50000,100)
for i in sampling:
    replica_0.append(replica[i][24])
    replica_1.append(replica[i][29])
    replica_2.append(replica[i][34])
    replica_3.append(replica[i][39])

#plot the trajectory of each replica
plt.plot(time,replica_0)
plt.title('Replica 1')
plt.xlabel('Time(ps)')
plt.ylabel('Temperature index')
#plt.savefig('Replica 1.pdf')
plt.show()

plt.plot(time,replica_1,color='r')
plt.title('Replica 2')
plt.xlabel('Time(ps)')
plt.ylabel('Temperature index')
#plt.savefig('Replica 2.pdf')
plt.show()

plt.plot(time,replica_2,color='g')
plt.title('Replica 3')
plt.xlabel('Time(ps)')
plt.ylabel('Temperature index')
#plt.savefig('Replica 3.pdf')
plt.show()

plt.plot(time,replica_3,color='k')
plt.title('Replica 4')
plt.xlabel('Time(ps)')
plt.ylabel('Temperature index')
#plt.savefig('Replica 4.pdf')
plt.show()
```

```python
#histogram of each replica potential
#load data from potential xvg files
with open("T300_potential.xvg") as f:
    T300_p = f.read()
    T300_p = T300_p.split('\n')
T300_potential = []
for line in T300_p[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    T300_potential.append(float(x[1]))
    
with open("T350_potential.xvg") as f:
    T350_p = f.read()
    T350_p = T350_p.split('\n')
T350_potential = []
for line in T350_p[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    T350_potential.append(float(x[1]))
    
with open("T400_potential.xvg") as f:
    T400_p = f.read()
    T400_p = T400_p.split('\n')
T400_potential = []
for line in T400_p[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    T400_potential.append(float(x[1]))

with open("T450_potential.xvg") as f:
    T450_p = f.read()
    T450_p = T450_p.split('\n')
T450_potential = []
for line in T450_p[:-1]:
    if "#" in line:
        continue
    if "@" in line:
        continue
    x = line.strip().split(' ')
    while '' in x:
        x.remove('')
    T450_potential.append(float(x[1]))
    
plt.hist(T300_potential,bins=20,alpha = 0.5,label='replica 1')
plt.hist(T350_potential,bins=20,alpha = 0.5,label='replica 2')
plt.hist(T400_potential,bins=20,alpha = 0.5,label='replica 3')
plt.hist(T450_potential,bins=20,alpha = 0.5,label='replica 4')
plt.title('Potential energy distribution')
plt.xlabel('Potential energy (KJ/mol)')
plt.ylabel('Counts')
plt.legend()
#plt.savefig('Potential energy distribution.pdf')
plt.show()
```
